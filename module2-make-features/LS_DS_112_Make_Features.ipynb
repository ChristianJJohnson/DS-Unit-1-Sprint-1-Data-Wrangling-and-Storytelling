## Work with strings

For machine learning, we usually want to replace strings with numbers.

We can get info about which columns have a datatype of "object" (strings)

```python
# Here I can see that a column I would think was a float or interger has a dtype of object
df['int_rate'].head()
```

```python
# Get the value of the first cell in this column
int_rate0 = df.loc[0, 'int_rate']
print(int_rate0)
int_rate0
```

```python
# The value has a percent sign and a extra space we are gonna want to get rid of
int_rate0 = int_rate0.strip().strip('%')
```

```python
# We can use that to then turn it into a float
float(int_rate0)
```

```python
print(type(float(int_rate0)))
```

### Convert `int_rate`

Define a function to remove percent signs from strings and convert to floats

```python
def int_rate_2_float(mystring):
  return float(mystring.strip().strip('%'))
```

```python
int_rate_2_float(int_rate0)
```

Apply the function to the `int_rate` column

```python
# apply your function to the whole columns
df['int_rate_float']=df['int_rate'].apply(int_rate_2_float)
df.columns
```

```python
# compare the old and new columns
df[['int_rate', 'int_rate_float']].head()
```

```python
# confirm that we can now employ math on this
df['int_rate_double'] = df['int_rate_float']*2
df[['int_rate', 'int_rate_float', 'int_rate_double']].head()
```

Clean the title and handle missing values

```python
# Function to clean up titles
def clean_title(title):
  if isinstance(title, str):
    return title.strip().title()
  else:
    return 'Unknown'

examples=['dog catcher', ' Hobo', ' Project Manager', np.NaN]

for item in examples:
  print(clean_title(item))
```

```python
# apply it to the entire dataframe
df['emp_title_clean']=df['emp_title'].apply(clean_title)
df.columns
```

```python
# compare
df[['emp_title', 'emp_title_clean']].sample(6)
```
